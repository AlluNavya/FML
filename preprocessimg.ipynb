```python
data['Age']=data['Age'].fillna(30)
```

```python
data['Age'].isnull().sum()
```

```python
print(data['Age'].mean() ,data['Age'].std(), data['Age'].min() ,data['Age'].max())
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x)
```

```python
print(data['Age'].mean() ,data['Age'].std(), data['Age'].min() ,data['Age'].max())
```

```python
y=scaler.transform(x)
```

```python
y.mean(),y.std(),y.min(),y.max()
```
